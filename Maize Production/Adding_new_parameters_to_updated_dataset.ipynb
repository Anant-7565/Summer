```python
df = pd.read_excel('updated_case_file.xlsx')
```

```python
df_new_data = pd.read_excel('Agriculture_KAG_2016_17.xlsx')
```

```python
df_new_data['Total_Production'] = df_new_data['TotalCerealsandMinorMillets_Production_230'] + df_new_data['TotalOilSeeds_Production_249'] + df_new_data['TotalFoodGrains_240'] + df_new_data['TotalGourdVarietyVegetables_276']
df_new_data['Ratio'] = df_new_data['Maize_Production_226']/df_new_data['Total_Production']
```

```python
df['Sowing Seeds'] = df_new_data['SowingSeedsDistributed_Maize_320'] * df_new_data['Ratio']
```

```python
df['Medium Landholdings'] = df_new_data['SemiMediumAgricultureLandHoldings_TotalArea_1077'] * df_new_data['Ratio']
```

```python
df['Small Landholdings'] = df_new_data['SmallAgricultureLandHoldings_TotalArea_1045'] * df_new_data['Ratio']
```

```python
df['Institutions Area'] = df_new_data['MediumAgricultureLandHoldings_TotalInstitutionsArea_1108'] * df_new_data['Ratio']
```

```python
remnants = ['AreaUnderCereals_Maize_193', 'SowingSeedsDistributed_Maize_320', 'District', 'Maize_Production_226']
temp = pd.DataFrame()
for column in df.columns:
    if( column not in remnants ):
        temp[column] = df[column]*df_new_data['Ratio']
    else:
        temp[column] = df[column]
temp
```

```python
temp.to_csv('maize_production_FCR_added4params.csv')
```
