```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('groundnutProductionCaseFile_KAG201617.xlsx')
```

```python
df.columns
```

```python
df_new_data = pd.read_excel('Agriculture_KAG_2016_17.xlsx')
```

```python
df_new_data['Total_Production'] = df_new_data['TotalCerealsandMinorMillets_Production_230'] + df_new_data['TotalOilSeeds_Production_249'] + df_new_data['TotalFoodGrains_240'] + df_new_data['TotalGourdVarietyVegetables_276']
df_new_data['Ratio'] = df_new_data['Groundnut_Production_241']/df_new_data['Total_Production']
#print(df['Maize_Production_226'])
```

```python
df_new_data['Ratio']
```

```python
df['TotalNPK_315']
```

```python
df_new_data['Ratio'] * df['TotalNPK_315']
```

```python
df = df.drop('District', axis=1)
#df = df.drop('Unnamed: 0', axis=1)
params = ['Groundnut_Production_241','AreaUnderOilSeeds_GroundNut_210','Rural_AnthyodayaBPLCardHolders_32','Total_AnthyodayaBPLCardHolders_33',
         'RuralBPLPriorityCardHoldersAkshayaWithoutCylinder_38',
       'TotalBPLPriorityCardHoldersAkshayaWithoutCylinder_39',
       'LivestockCensus_2012_Buffaloes_355', 'LivestockCensus_2012_Sheep_359',
       'LivestockCensus_2012_Goats_360','SowingSeedsDistributed_Groundnut_328']
for column_name in df.columns:
    print(column_name)
    if( column_name not in params ):
        df[column_name] = df[column_name] * df_new_data['Ratio'] 
df
```

```python
df.to_csv('FCR_groundnutProductionCaseFile_KAG201617.csv')
```

```python
X = df.drop(columns=['Groundnut_Production_241'])
y = df['Groundnut_Production_241']
```

```python
n = len(X.columns)

# Create subplots
fig, axs = plt.subplots(n, 1, figsize=(8, 5 * n))

# Plot each x column
for i, x in enumerate(X.columns):
    axs[i].scatter(df[x], y, label=f'Scatter: {x}', color='b')
    axs[i].set_title(f'Plot for {x} vs {y.name}')
    axs[i].set_xlabel(x)
    #axs[i].set_ylim([0, 100000])
    #axs[i].set_xlim([0, 10000])
    axs[i].set_ylabel(y.name)
    #axs[i].legend()

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
params = ['NetAreaIrrigatedUnderDifferentSources_TubeWells_GrossIrrigatedArea_181',
       'NetAreaIrrigatedUnderDifferentSources_TotalGrossIrrigatedArea_188',
       'NetAreaIrrigatedUnderDifferentSources_TotalNetIrrigatedArea_189',
       'AreaUnderOilSeeds_GroundNut_210','MotorVehicles_Tractors_514',
       'MotorVehicles_Trailers_515','LivestockCensus_2012_Sheep_359','SemiMediumAgricultureLandHoldings_TotalNumber_1073',
       'SemiMediumAgricultureLandHoldings_TotalArea_1077',
       'MediumAgricultureLandHoldings_TotalNumber_1105',
       'MediumAgricultureLandHoldings_TotalArea_1109',
       'LargeAgricultureLandHoldings_TotalNumber_1137',
       'LargeAgricultureLandHoldings_TotalArea_1141', 'TotalNPK_315','TotalAgricultureLoan','Groundnut_Production_241']
#df = df.drop('Unnamed: 0', axis=1)
save_df = pd.DataFrame()
for column_name in df.columns:
    print(column_name)
    if( column_name in params ):
        save_df[column_name] = df[column_name]
save_df['District'] = df_new_data['District']
save_df.columns
```
